#**Atividade AI**

##1. Carregando Dados

```python
import pandas as pd
import seaborn as sns
```

```python
from seaborn import load_dataset
df = load_dataset('titanic')
df
```

##2. Ajustando colunas e retirando os valores NaN
- As colunas 'class' e 'plass', exibem a mesma informação para o propósito em questão.

```python
df = df.drop(columns = ['class','deck','alive','adult_male','embark_town','sex'])
df = df.dropna()
```

```python
df
```

##3. Transformando os atributos categóricos com One-Hot Encoder

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
c_trans = make_column_transformer((OneHotEncoder(), ['alone','who','embarked']), remainder='passthrough')
```

```python
final_df = pd.DataFrame(data=c_trans.fit_transform(df), columns=c_trans.get_feature_names_out())
final_df = final_df.drop('onehotencoder__alone_False', axis=1)
```

```python
final_df.columns = ['Alone','Child','Man','Woman','Cherboug_embarked','Queenstown_embarked','Southampton_embarked','Survived','Class','Age','Sibsp','Parch','Fare']
final_df
```

##4. Separando os dados de Treinamento e teste

- A coluna Target é a 'Survived'.

```python
y = final_df['Survived']
X = final_df.drop(labels=['Survived'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y)
```

##5. Treinando duas árvores sob parâmetros diferentes

```python
from sklearn.tree import DecisionTreeClassifier

model1 = DecisionTreeClassifier(criterion= 'gini', max_depth=4)
model2 = DecisionTreeClassifier(criterion='entropy', max_depth=2)

model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
```

##6. Treinando dois kNN sob parâmetros diferentes



```python
from sklearn.neighbors import KNeighborsClassifier

knn1 = KNeighborsClassifier(n_neighbors=5, metric='euclidean', algorithm='brute')
knn2 = KNeighborsClassifier(n_neighbors=5, metric='minkowski', algorithm='kd_tree')

knn1.fit(X_train, y_train)
knn2.fit(X_train, y_train)
```

```python
from sklearn import metrics

tree_result1 = model1.predict(X_test)
tree_result2 = model2.predict(X_test)
knn_result1 = knn1.predict(X_test)
knn_result2 = knn2.predict(X_test)

acc1 = metrics.accuracy_score(tree_result1, y_test)
show1 = round(acc1 * 100)
print("DecisionTreeClassifier(criterion= 'gini', max_depth=4) -> {}%".format(format(show1)))

acc2 = metrics.accuracy_score(tree_result2, y_test)
show2 = round(acc2 * 100)
print("DecisionTreeClassifier(criterion= 'entropy', max_depth=2) -> {}%".format(format(show2)))

acc3 = metrics.accuracy_score(knn_result1,y_test)
show3 = round(acc3 * 100)
print("KNeighborsClassifier(n_neighbors=5, metric='euclidean', algorithm='brute') -> {}%".format(format(show3)))

acc4 = metrics.accuracy_score(knn_result2,y_test)
show4 = round(acc4 * 100)
print("KNeighborsClassifier(n_neighbors=5, metric='minkowski', algorithm='kd_tree') -> {}%".format(format(show4)))
```
